## Do students that majored in subjects that were majority female make more money?
There seems to be a negative correlation between median wage and percentage of women in the major

Women tend to choose majors where median income is lower. 
This could explain the alleged "gender gap". 

It would be interesting to explore if women earn less or more whithin specific majors.
